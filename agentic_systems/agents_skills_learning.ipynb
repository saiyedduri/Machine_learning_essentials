# What are Agent Skills?

The ability to create,manage and share skills by extending the capabilties of the LLM and the code itself.

These skills are modular capabiltoes that extend the functionality through organized folders containing instructions, scripts, and resources.

Each skill consists of skills.md file with instructions that claude reads when relevant, plus optional supporting files like scripts and templates.

Skills are model invoked-decided by claude/LLM to use them based on your request and skills description. 
This is different from the slash commands, which are user invoked.

Skills extend capabilyies of the LLM, transforming general purpose agents into specialized agents that fits your needs.

Bigger Picture Benifits:
- Agentic skills learning extend the capabilities of the specific workflow by invoking the skills when ever required.

- This would reduce the repetative prompting.

- Compose multiple skills for complex tasks.

Agentic skills solve this problem:
LLM's are smart and general-purpose, but real work often requires specialized knowledge—like how to fill out a PDF form, create a PowerPoint, or follow your company's specific processes. You could put all this information in the prompt every time, but that's wasteful and doesn't scale

LLM only loads what it needs, when it needs it.

# Objective of agentic systems design:

A the model capabilities improve, we can now build general purpose agents that can interact with the computing environments, databases, codebases to  accomplish tasks are complex tasks while being composable, scalable, and portable.

Composable means you can combine smaller, modular pieces of expertise together like building blocks. 

Portable means the expertise can be moved and reused across different agents, platforms, or environments without being locked to one specific system.

A **Skill** is basically a folder containing instructions and tools that teach Claude how to do something specific.

Think of it like this:

> **Analogy:** Imagine you're training a new employee. Instead of explaining everything from scratch each day, you give them an employee handbook. When they need to do a specific task (like expense reports), they flip to that chapter. Skills work the same way for Claude.

## How a Skill is Structured

At minimum, a skill is just a folder with one file: `SKILL.md`

```
pdf-skill/
  └── SKILL.md
```

The `SKILL.md` file has two parts:

1. **Metadata (name + description)** — A brief summary so Claude knows *when* to use this skill
2. **Instructions** — The actual how-to guide

For more complex skills, you can add extra files:

```
pdf-skill/
  ├── SKILL.md          (main instructions)
  ├── forms.md          (how to fill out forms)
  ├── reference.md      (technical details)
  └── extract_fields.py (code Claude can run)
```

---

## The Key Idea: Progressive Disclosure

This is the most important concept. It means **Claude only loads what it needs, when it needs it.**

| Level | What Claude Sees | When |
|-------|------------------|------|
| 1 | Just the name & description of all skills | Always (at startup) |
| 2 | The full `SKILL.md` content | When Claude decides the skill is relevant |
| 3 | Additional files (like `forms.md`) | Only when that specific subtopic comes up |

> **Analogy:** It's like a textbook. You see the table of contents first, then open the chapter you need, then maybe check the appendix for extra details. You don't read the whole book every time.

**Why this matters:** Claude has a limited "context window" (how much it can hold in memory at once). Progressive disclosure keeps it efficient.

---

## Skills Can Include Code

Some tasks are better done with traditional code than with AI "thinking." For example:

- Sorting a list → Code is faster and cheaper
- Extracting form fields from a PDF → Deterministic code is more reliable

A skill can include Python scripts that Claude runs when needed, rather than trying to figure things out from scratch.

## Real Example: The PDF Skill

The article uses a PDF skill as an example:

1. User asks: *"Fill out this tax form"*
2. Claude sees it has a "PDF skill" available (from the metadata)
3. Claude reads the main `SKILL.md` file
4. Claude sees it should also read `forms.md` for form-filling instructions
5. Claude runs a bundled Python script to extract the form fields
6. Claude fills out the form using the instructions

## Best Practices for Building Skills

| Tip | What It Means |
|-----|---------------|
| **Start with evaluation** | Watch where Claude fails, then build skills to fix those gaps |
| **Structure for scale** | Split big instructions into multiple files |
| **Think from Claude's perspective** | Write clear names and descriptions so Claude knows when to use the skill |
| **Iterate with Claude** | Ask Claude to help you write the skill based on what worked |

---

## The Big Picture

Skills make it possible to:

- **Compose** — Combine multiple skills for complex tasks
- **Scale** — Add lots of specialized knowledge without bloating every prompt
- **Share** — Package expertise into reusable folders anyone can use

It's a simple idea (just folders and markdown files!) that enables powerful customization.
