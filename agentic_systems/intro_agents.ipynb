# Objective of Agents:
Autonomous goal oriented agents plan and act over multiple steps to solve complex problems.

This primarily involves 3 primary mechanisms:
1. model: Reasoning engine that manages the context window. Information comes from memory,files,prompt. Model decides the thought for next step by understanding the input and past memory and goals from the context window.

2. Orchestration: Governs the interaction of the models with tools . 
    This involves
    1. Plans and manages the operational loop, keeping the memory/state.
    2. Executing the chain of thought and assignment the tool for the thought
Mainly, it decides when the Model should think and when Tools should act.
    3. If needed, we can also define the agent persona.
    
3. Tools: booking apis, extraction tools, reading and writing tools.


---

## The 4 Components

| Component | Analogy | What It Does |
|-----------|---------|--------------|
| **Model** |  Brain | Reasons, decides, plans |
| **Tools** |  Hands | Connects to outside world (APIs, databases, code) |
| **Orchestration Layer** |  Nervous System | Manages the loop, memory, and coordination |
| **Deployment** |  Body & Legs | Makes it accessible, scalable, and reliable |

#  Agent Types (Reasoning Patterns)

Different agent architectures for different problem types:

| Agent Type | How It Works | Best For |
|------------|--------------|----------|
| **Reflex** | Simple IF-THEN rules, no planning | Simple, predictable tasks |
| **ReAct** | Interleaves Reasoning + Acting in a loop | Most general-purpose tasks |
| **Planner-Executor** | First plans all steps, then executes | Well-defined multi-step tasks |
| **Query-Decomposition** | Breaks complex queries into sub-questions | Research, complex Q&A |
| **Reflection** | Self-evaluates output and improves | Quality-critical tasks |
| **Deep Research** | Multi-step research with synthesis | In-depth analysis |

### ReAct Pattern (Most Common)
```
Thought â†’ Action â†’ Observation â†’ Thought â†’ Action â†’ ... â†’ Final Answer
```

### Planner-Executor Pattern
```
Plan All Steps â†’ Execute Step 1 â†’ Execute Step 2 â†’ ... â†’ Compile Results
```

### Reflection Pattern
```
Generate Response â†’ Critique Response â†’ Improve Response â†’ (repeat if needed)
```

# Different levels of Agents: Define the complexity for Agents

Level0: Just the trained language model(will not be able to tell info on the things that it was not trained on. Ex: todays morning cricket score)

Level1: Language model is connected to tools such as search apis so that it have realtime awareness.

Level2: Usage of context engineering for strategically solving multi step process.
Ex: Find the coffee between the 2 locations. This involves:
1. Use maps tool to find the midpoint of the both location
2. then use another tool to find the shop that are near to the location with the rating more than 4.0.

Level3: Collaborative multi agents systems:
project manager agent: delegates market research to market research agent.
market research agent: receives the request from project manager agent and executes multistep plan that create its own response plan and execute the plan.

Level4: Self-evolving system:
Systems can identify the gaps in its own capabilities and adapt, autonomously create and extend its own toolkit.
Ex: Can build the sentimental analysis for research.

# Memory:
Short-term memory: Scratchpad for the current task. The running history of acts and observation is stored in the short term memory.

Long-term memory: Long-term memory persists across session.
Usually a RAG system where messages are retrieved from the vector databases.

#  Memory Systems Deep Dive

## Context Window Management
The context window is LIMITED. Strategies to manage it:
- **Summarization**: Compress older messages
- **Selective Retrieval**: Only pull relevant information
- **Sliding Window**: Keep only recent N messages

# Learning in Agentic Systems

## Nonparametric Learning (No Model Retraining)

| Method | How It Works |
|--------|-------------|
| **Exemplar Learning** | Store examples in memory, retrieve similar ones for few-shot prompting |
| **Reflexion** | Agent reflects on failures, stores insights, improves next attempt |
| **Experiential Learning** | Log successful strategies, retrieve them for similar future tasks |

### Reflexion Pattern
```
Attempt Task â†’ Fail â†’ Reflect ("What went wrong?") â†’ Store Insight â†’ Retry with Insight
```

## Parametric Learning (Model Retraining)

| Method | Description |
|--------|-------------|
| **SFT (Supervised Fine-Tuning)** | Train on labeled input-output pairs |
| **DPO (Direct Preference Optimization)** | Train on preference pairs (good vs bad responses) |
| **RLHF** | Reinforcement Learning from Human Feedback |
| **RLVR** | Reinforcement Learning with Verifiable Rewards (for tasks with clear right/wrong) |

## When to Use Multiple Agents?

 **Use Multiple Agents:**
- Different specialized expertise needed
- Tasks can be parallelized
- Complex workflows with handoffs
- Actor-critic quality control

 **Stick to Single Agent:**
- Well-defined, bounded tasks
- Need for simplicity and debuggability
- Latency-critical applications
- Cost-sensitive scenarios

#  Security Deep Dive

## Threat Vectors

| Attack Type | Description | Defense |
|-------------|-------------|----------|
| **Prompt Injection** | Malicious instructions in user input | Input sanitization, instruction hierarchy |
| **Data Poisoning** | Corrupted training/memory data | Data validation, provenance tracking |
| **Model Extraction** | Stealing model behavior | Rate limiting, output filtering |
| **Jailbreaking** | Bypassing safety guidelines | Multi-layer guardrails, red teaming |

## Defense Strategies

1. **Input/Output Filtering** - Block malicious patterns
2. **Anomaly Detection** - Flag unusual behavior
3. **Access Control** - RBAC/ABAC for tools and data
4. **Audit Logging** - Track all agent actions
5. **Red Teaming** - Proactively test for vulnerabilities

## Data Protection

- **Encryption** - At rest and in transit
- **PII Anonymization** - Remove/mask sensitive data
- **Data Provenance** - Track data origin and transformations
