# Objective of Agents:
Automatous goal oriented agents plan and act over multiple steps to solve complex problems.

This primarily involves 3 primary mechanisms:
1. model: Reasoning engine that manages the context window. Information comes from memory,files,prompt. Model decides the thought for next step by understanding the input and past memory and goals from the context window.

2. Orchestatration: Governers the interation of the models with tools . 
    This involves
    1. Plans and manages the operatoinal loop, keeping the memory/state.
    2. Executing the chain of thought and assignment the tool for the thought
Mainly, it decides when the Model should think and when Tools should act.
    3. If needed, we can also define the agent persona.
    
3. Tools: booking apis, extraction tools, reading and writing tools.


---

## The 4 Components

| Component | Analogy | What It Does |
|-----------|---------|--------------|
| **Model** |  Brain | Reasons, decides, plans |
| **Tools** |  Hands | Connects to outside world (APIs, databases, code) |
| **Orchestration Layer** |  Nervous System | Manages the loop, memory, and coordination |
| **Deployment** |  Body & Legs | Makes it accessible, scalable, and reliable |

# Different levels of Agents: Define the complexity for Agents

Level0: Just the trained language model(will not be able to tell info on the things that it was not trained on. Es: todays morning cricket score)

Level1: Language model is connected to tools such as search apis sothat it have realtime awaremeness.

Level2: Usage of context engineering for strategically solving multi step process.
Ex: Find the cofee between the 2 locations. This involves:
1. Use maps tool to find the midpoint of the both location
2. then use another tool to find the shop that are near to the location with the rating more than 4.0.

Level3: Collaborative multi agents systems:
project manager agent: deligates market researach to market research agent.
market research agent: recieves the request from project manager agent and executes multistep plan that create its own response plan and execute the plan.

Level4: Self-evolving system:
Systems can identify the gaps in its own capabilities and adapt, autonomusly create and extend its own toolkit.
Ex: Can build the sentimental analysis for research.

# Model Routing:
  Sending different tasks for different models.
  For ex: Sending complex tasks to latest/advanced model and basic task to fundamental models.
  
  This would reduce the traffic and cost(better resource allocation).
