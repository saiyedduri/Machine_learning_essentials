# Image: 

An image is a function u:Ω→F that maps each point in a domain 
Ω (the pixel positions) to a value in a color spacce F (like grayscale or RGB)

- If $\Omega \subset \mathbb{Z}^d$, the image is discrete(digital, made of pixels).

- If $\Omega \subset \mathbb{R}^d$, the image is continuous (ideal or analog).

- In 2D, a discrete image is a matrix, where $u(i, j)$ gives the color at row $i$ and column $j$





Discrete image is what we normally use on computers- its made up of pixels, arranged in a grid.

- Efficiently stored in a computer:
    Each pixel has a value (eg., color or brightness), and computers store these values in arrays or matrics- which is fast and memort-friendly.
- Many image operations are simple matrix operations:
    Since the image is a matrix, you can apply filters, adjust brightness, blur, etc., using matrix addition multipication, and convolution.
- Operations like scaling and rotating are not perfectly lossless

When you resize or rotate a discrete image:
 - You often need to estimate values bw pixels 
 - This can cause blurring or loss of detail.
 - It's not mathematically exact because the image is limited to fixed pixels.

# Continuous image:

A continuous image is a smooth mathematical function- like a formula that gives you  a color at any function(not just on a pixel grid)
- Used as the mathematical models behind discrete images
 Think of a continuous image as the ideal, real-world scene. When we take a photo or scan, we sample this continuous image into pixels to make a discrete one.

- Need function basis to be stored in a computer:
    Computers can't store infinite values, so to represent a continuous image.
    - We approximate it using known mathematical building blocks(called  function basis), like:
        - Fourier series
        - Wavelets
        - Polynomials

- The computer then stores just the coefficients of these functions.

# How projectors or screens work?
Input: R,G,B
- LEDs of different color
- Screens use LEDs or backlights with color filters to emit red, green, and blue light.
- These lights mix together to create the colors we see on screen.

To make RGB values comparable across devices, we need a color profile or transformation — that’s where color management and absolute color spaces (like CIELAB) come in.

# File formats for images:
 JPEG: Joint Photographic Experts group
 suited for photographs, uses FFT for compression.

 BMP: Windows Bitmap

 proprietary formats of the camera production companies, saving data directly after quantization.
