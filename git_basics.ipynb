## 1. 🤔 What is Git?

Git tracks changes to files over time. Think of it as a **time machine** for your code.

```
Without Git                  With Git
───────────                  ────────
project_v1.zip               One folder, full history
project_v2.zip         →     Easy collaboration
project_final.zip            Rollback anytime
project_FINAL2.zip           Work on features in parallel
```

### Key Concepts
| Term | Meaning |
|---|---|
| **Repository** | A folder tracked by Git |
| **Commit** | A saved snapshot of changes |
| **Branch** | An independent line of development |
| **Remote** | A hosted copy on GitHub/GitLab |
| **HEAD** | Pointer to your current position |
| **Staging Area** | Where you prepare files before committing |

## 4. 📸 Tracking Changes — add & commit

Git has a **3-step workflow**:

```
Working Directory   →   Staging Area   →   Repository
  (edit files)          (git add)          (git commit)
```

### Step 1 — Create or edit files

````python
os.chdir("/tmp/my-project")

# Create a sample file
with open("README.md", "w") as f:
    f.write("# My Project\nThis is my first Git project!\n")

with open("app.py", "w") as f:
    f.write("print('Hello, Git!')\n")

print("✅ Files created: README.md, app.py")
````

### Step 2 — Check status

````python
# See what Git sees
!git status
````

### Step 3 — Stage files (git add)

````python
# Stage a specific file
!git add README.md

# Stage all files at once
!git add .

# Check status again — files are now staged
!git status
````

### Step 4 — Commit (git commit)

> A commit is like a **save point** — a permanent snapshot you can return to.

````python
!git commit -m "Initial commit: add README and app.py"
print("\n✅ Commit saved!")
````

### 💡 Common git add variants
```bash
git add file.txt         # Stage one file
git add .                # Stage all changes
git add *.py             # Stage all Python files
git add -p               # Interactively stage chunks
```

### 💡 Good commit message tips
```bash
# ✅ Good
git commit -m "Fix login bug for OAuth users"
git commit -m "Add payment processing module"

# ❌ Bad
git commit -m "fix"
git commit -m "changes"
```

## 6. 🌿 Branching

Branches let you work on features without affecting the main codebase.

```
main:    A ── B ── C
                    \
feature:             D ── E   ← you work here safely
```

````python
# First, commit our current change
!git commit -m "Update README with features section"

# List all branches
!git branch
````

````python
# Create a new branch
!git branch feature-login

# Switch to it
!git checkout feature-login

# OR do both in one command (shortcut)
# git checkout -b feature-login

print("\nCurrent branch:")
!git branch
````

````python
# Add a feature on the new branch
with open("login.py", "w") as f:
    f.write("def login(user, password):\n    return user == 'admin' and password == 'secret'\n")

!git add login.py
!git commit -m "Add login module"

print("\n✅ Feature committed on feature-login branch")
````

````python
# Switch back to main — notice login.py is gone!
!git checkout main

import os
print("Files on main:", os.listdir("."))
````

### Branch Commands Summary
```bash
git branch                  # List branches
git branch feature-x        # Create branch
git checkout feature-x      # Switch to branch
git checkout -b feature-x   # Create + switch
git branch -d feature-x     # Delete branch (safe)
git branch -D feature-x     # Force delete
git branch -m old new       # Rename branch
```

## 7. 🔀 Merging

Merging combines a branch back into another (usually `main`).

````python
# We're on main — merge the feature branch in
!git merge feature-login

print("\nFiles after merge:", os.listdir("."))
print("\n✅ login.py is now in main!")
````

### Handling Merge Conflicts

When two branches edit the same part of a file, Git can't auto-merge:

```
<<<<<<< HEAD
your changes on main
=======
their changes on feature branch
>>>>>>> feature-login
```

**Resolution steps:**
1. Open the conflicting file
2. Choose which code to keep (or combine both)
3. Remove the conflict markers (`<<<<<<<`, `=======`, `>>>>>>>`)
4. `git add <file>`
5. `git commit`

### Merge vs Rebase
| | Merge | Rebase |
|---|---|---|
| History | Preserves all commits | Rewrites history (cleaner) |
| Safety | Safe for shared branches | Avoid on shared branches |
| Use when | Finishing a feature PR | Updating a local branch |

## 8. ☁️ Remote Repositories (push & pull)

A **remote** is a hosted copy of your repo (GitHub, GitLab, Bitbucket).

```
LOCAL                        REMOTE (GitHub)
──────                       ───────────────
  │  ◄──── git pull ────────│   (download)
  │  ──── git push ─────────►│   (upload)
```

```bash
# Link your local repo to a remote
git remote add origin https://github.com/username/repo.git

# View linked remotes
git remote -v

# Push your code to GitHub
git push -u origin main       # First time (-u sets upstream)
git push                      # After that, just this

# Push a new branch
git push origin feature-login

# Download latest changes
git pull origin main

# Fetch without merging (safer)
git fetch origin
```

### Typical Remote Workflow
```bash
# Start of day
git pull origin main                    # 1. Get latest

# Work on feature
git checkout -b feature-dashboard       # 2. New branch
# ...make changes...
git add . && git commit -m "Add dash"   # 3. Commit
git push -u origin feature-dashboard   # 4. Push branch

# Open Pull Request on GitHub           # 5. PR & review
git checkout main && git pull           # 6. Sync after merge
```

## 9. ↩️ Undoing Changes

Git lets you undo at every stage.

````python
# Scenario: We made a bad change
with open("app.py", "a") as f:
    f.write("\n# This is a mistake\nbad_code_here()\n")

print("Bad change made to app.py")
!git status
````

````python
# UNDO 1: Discard changes in working directory (not staged)
!git restore app.py
print("\n✅ app.py restored to last commit state")
!git status
````

````python
# UNDO 2: Unstage a file (if you already did git add)
with open("app.py", "a") as f:
    f.write("\nbad_code()\n")

!git add app.py
print("Staged the bad file...")

!git restore --staged app.py
print("\n✅ Unstaged app.py (changes still in working directory)")
````

### Undoing Commits

```bash
# Undo last commit — keep changes staged
git reset --soft HEAD~1

# Undo last commit — keep changes unstaged
git reset --mixed HEAD~1

# Undo last commit — discard changes entirely ⚠️
git reset --hard HEAD~1

# Safely undo a pushed commit (creates new undo commit)
git revert HEAD

# Revert a specific commit by hash
git revert abc1234
```

### When to use what?
| Situation | Command |
|---|---|
| Discard file changes | `git restore file.txt` |
| Unstage a file | `git restore --staged file.txt` |
| Undo last local commit | `git reset --soft HEAD~1` |
| Undo pushed commit safely | `git revert HEAD` |

## 10. 🙈 .gitignore

Tell Git to ignore specific files or folders (secrets, dependencies, build outputs).

````python
# Create a .gitignore file
gitignore_content = """# Python
__pycache__/
*.pyc
*.pyo
.env
venv/

# Node
node_modules/

# Build outputs
dist/
build/
*.log

# OS files
.DS_Store
Thumbs.db

# Secrets
*.env
secrets.json
"""

with open(".gitignore", "w") as f:
    f.write(gitignore_content)

print(".gitignore created:")
print(gitignore_content)
````

````python
# Create a file that should be ignored
with open(".env", "w") as f:
    f.write("SECRET_KEY=my-super-secret-key\n")

!git add .gitignore
!git commit -m "Add .gitignore"

# .env should NOT appear in git status
print("\ngit status (notice .env is ignored):")
!git status
````

## 🏆 Golden Rules of Git

1. **Commit often** — small, focused commits are easier to manage
2. **Pull before you push** — always sync before uploading
3. **Never force push to shared branches** — rewrites history for everyone
4. **Write meaningful commit messages** — your future self will thank you
5. **Use branches for everything** — keep `main` always stable
6. **Never commit secrets** — use `.gitignore` and environment variables
